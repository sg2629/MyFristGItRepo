**Exercise 3.** Hot-deck imputation replaces missing data with similar entries.  We will use a stochastic hot-deck imputation method, which replaces the number captured into slavery (tslavesd) with a random voyage from the same decade.  (If there is only one voyage in that decade then ignore that record.)  Construct a function that performs this random imputation method, and outputs a Series with this new imputed variable.  Calculate the total imputed tslavesd variable 100 times by performing this random imputation method 100 times.  Report the standard error, which is the standard deviation of the total imputed, along with the mean of the 100 trials.

````python
import random
def hot_deck (TAST_na):
    """
    According to the year, get the corresponding decade.
    Then, group by the decade and replace the missing values in each decade block.
    choice: non-NA values in column 'tslavesd'
    """
    TAST_na['decade']=(TAST_na['yearam']//10)*10  # create a new variable 'decade'
    gb=TAST_na.groupby('decade')
    tslavesd_new=[]
    for j in list(gb.groups.keys()):
        temp=gb.get_group(j)[['decade','tslavesd']].astype(float)
        choice=list(temp[temp['tslavesd'].notna()]['tslavesd'].astype(float))
        for index, row in temp.iterrows():
            if choice !=[] and pd.isnull(row[1]):
                temp.loc[index,['tslavesd']]=random.choice(choice) 
        tslavesd_new.append(temp['tslavesd'])
    return(pd.concat(tslavesd_new))
````

````python
total=np.array([np.sum(hot_deck(TAST_na)) for _ in range(100)])
np.mean(total)
````

````python
np.std(total)
````
